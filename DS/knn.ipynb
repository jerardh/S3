```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
x=iris.data
y=iris.target
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
model = KNeighborsClassifier(n_neighbors=5) # setting k=5
model.fit(x_train,y_train)
```

```python
y_prediction = model.predict(x_test)
print("Accuracy = ",metrics.accuracy_score(y_test,y_prediction)) # comparing expected and actual outputs
```

Predicting a new instance

```python
sample = [[2,2,2,2]] # 1 sample with 4 attribuites - petal length,petal width, sepal length, sepal width
pred = model.predict(sample)[0] # getting first prediction as the index of the flower type
print("New Sample belongs to  ",iris.target_names[pred])
```

```python
plt.subplot(1,2,1)
plt.scatter(x_test[:,0],x_test[:,1],c=y_prediction)
plt.title("Predicted classification")
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.subplot(1,2,2)
plt.scatter(x_test[:,0],x_test[:,1],c=y_test)
plt.title("Actual classification")
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.show()
```
