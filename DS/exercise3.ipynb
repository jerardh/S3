1.Draw a line from position (1,3) and (2,10) then to (6,12) and finally to position (18,20). Mark each point with a beatiful green color and set line colour to red and line style dotted.

```python
import matplotlib.pyplot as plt
import numpy as np
x_data = np.array([1,2,6,18])
y_data = np.array([3,10,12,20])
plt.plot(x_data,y_data,ls=":",marker=".",ms=10,mfc="green",mec="green",color="red")
```

Temperature in degree       Sales

3.Write a python program to draw a line using given axis values taken from a text file,
  with suitable labels in x and y axis and a title


```python

file=open("exercise3.3.txt")
x_data =np.array(list(map(int,file.readline().strip().split(","))))
y_data =np.array(list(map(int,file.readline().strip().split(","))))
file.close()
plt.title("Temperature Information")
plt.xlabel("Day")
plt.ylabel("Temperatrure")
plt.plot(x_data,y_data,ls="-",marker="*",mec="black",mfc="black")
```
